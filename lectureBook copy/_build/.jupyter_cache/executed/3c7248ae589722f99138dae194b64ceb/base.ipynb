```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.random.normal(170, 10, 250)
df = pd.DataFrame(x)

print(df)

plt.hist(df)
plt.title("Random df")
plt.ylabel("frequency")
plt.xlabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame({"Fruits": ["Apples", 'Bananas'], 
                   "Values": [12, 17]
                  },index = [1,2])
plt.bar(df["Fruits"],df["Values"])
plt.title("Random df")
plt.ylabel("Amount")
plt.xlabel("Fruits")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([35,25,20,20])
myLabels = ["Apples", "Banana", "Cherries", "Peaches"]
plt.pie(data, labels = myLabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as df
import numpy as np

x = df.DataFrame([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = df.DataFrame([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.title("x vs y")
plt.ylabel("y")
plt.xlabel("x")
plt.show()

```
